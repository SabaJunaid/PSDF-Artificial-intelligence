## Task 4 Product Discount Checker

```python
price = int(input("Enter product price: "))
membership = input("Are you a member? (yes/no): ")

if price > 5000 and membership == "yes":
    print("20% discount")
elif price > 5000 and membership == "no":
    print("10% discount")
else:
    print("No discount")
```

## Project : AI Recruitment Decision System

```python
name = input("Enter name: ")
age = int(input("Enter age: "))
cgpa = float(input("Enter CGPA: "))
skills = int(input("Enter skills count: "))
experience = int(input("Enter experience (years): "))
salary = int(input("Enter expected salary: "))
communication = int(input("Enter communication score (1-10): "))

# Fraud detection
if salary > 400000 and experience == 0:
    print("Suspicious Application")

# Reject conditions
elif age < 18 or cgpa < 2.7 or skills < 3:
    print("Rejected")

# Hiring levels
elif cgpa >= 3.3 and skills >= 7 and communication >= 7:
    print("Software Engineer")

elif cgpa >= 3.0 and skills >= 5 and experience >= 1:
    print("Junior Developer")

elif cgpa >= 2.7 and skills >= 3:
    print("Intern")

else:
    print("Not Eligible")
```
