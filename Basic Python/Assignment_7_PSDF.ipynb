Q1. Create a function that takes a list of numbers as argument and returns the sum of all even numbers present in the list.

```python
def sum_even(numbers):
    total = 0

    for num in numbers:
        if num % 2 == 0:
            total += num

    return total

nums = [10, 15, 20, 25, 30]
print("Sum of even numbers:", sum_even(nums))
```

Q2. Write a program that takes student marks as input, stores them in a list, and uses a function to calculate the average marks and grade

```python
def calculate_result(marks):
    average = sum(marks) / len(marks)

    if average >= 90:
        grade = "A+"
    elif average >= 80:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 50:
        grade = "D"
    else:
        grade = "F"

    return average, grade

marks = []

for i in range(5):
    mark = float(input(f"Enter marks of subject {i+1}: "))
    marks.append(mark)

avg, grade = calculate_result(marks)

print("Average Marks:", avg)
print("Grade:", grade)
```

Q3. Create a function that accepts a string and returns the number of uppercase letters, lowercase letters, and digits present in it

```python
def count_characters(text):
    upper = 0
    lower = 0
    digits = 0

    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1
        elif ch.isdigit():
            digits += 1

    return upper, lower, digits

string = input("Enter a string: ")

u, l, d = count_characters(string)

print("Uppercase Letters:", u)
print("Lowercase Letters:", l)
print("Digits:", d)
```

Q5. Create a menu-driven calculator using functions and loops that performs addition, subtraction, multiplication, and division

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Division by zero is not allowed"
    return a / b

while True:
    print("\nCalculator Menu")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "5":
        print(""Calculator Closed"")
        break

    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    if choice == "1":
        print("Result:", add(a, b))
    elif choice == "2":
        print("Result:", subtract(a, b))
    elif choice == "3":
        print("Result:", multiply(a, b))
    elif choice == "4":
        print("Result:", divide(a, b))
    else:
        print("Invalid Choice")
```

Q17. Create a banking system using functions where users can deposit, withdraw, and check balance. Prevent withdrawal if balance is insufficient.

```python
balance = 0

def deposit():
    global balance
    amount = float(input("Enter deposit amount: "))
    if amount > 0:
        balance += amount
        print("Deposited successfully")
    else:
        print("Invalid amount")

def withdraw():
    global balance
    amount = float(input("Enter withdraw amount: "))
    if amount > balance:
        print("Insufficient balance")
    elif amount <= 0:
        print("Invalid amount")
    else:
        balance -= amount
        print("Withdraw successful")

def check_balance():
    print("Current Balance:", balance)

while True:
    print("\n--- BANK SYSTEM ---")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        deposit()
    elif choice == "2":
        withdraw()
    elif choice == "3":
        check_balance()
    elif choice == "4":
        print("Exiting system")
        break
    else:
        print("Invalid choice")
```

Q18. Write a program that takes a sentence from the user and uses functions to count vowels, consonants, spaces, and digits.

```python
def count_chars(sentence):
    vowels = consonants = spaces = digits = 0

    for ch in sentence:
        if ch.isdigit():
            digits += 1
        elif ch == " ":
            spaces += 1
        elif ch.lower() in "aeiou":
            vowels += 1
        elif ch.isalpha():
            consonants += 1

    return vowels, consonants, spaces, digits

text = input("Enter sentence: ")

v, c, s, d = count_chars(text)

print("Vowels:", v)
print("Consonants:", c)
print("Spaces:", s)
print("Digits:", d)
```

Q25. Sum of prime numbers in tuple

```python
def is_prime(num):
    if num < 2:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False
    return True

def sum_primes(numbers):
    total = 0
    for num in numbers:
        if is_prime(num):
            total += num
    return total

nums = (2, 3, 4, 5, 6, 7, 10)

print("Sum of primes:", sum_primes(nums))
```

Q27. Password validation system

```python
def check_password(password):
    has_upper = False
    has_digit = False

    if len(password) < 8:
        return "Password must be at least 8 characters"

    for ch in password:
        if ch.isupper():
            has_upper = True
        if ch.isdigit():
            has_digit = True

    if not has_upper:
        return "Password must contain uppercase letter"
    if not has_digit:
        return "Password must contain a digit"

    return "Password is valid"

password = input("Enter password: ")

print(check_password(password))
```
