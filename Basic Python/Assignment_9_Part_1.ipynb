## 1. How does an object call methods of a class?

When you create an object, Python internally passes that object as the first argument (self) to the method.

```python
class A:
    def show(self):
        print("Hello")

obj = A()
obj.show()

"""A.show(obj) # happens internally
So, the object is just a reference
that tells the method which data it should work on."""
```

## 2. Runtime Type vs Reference Type

Reference type (static type): The type of variable reference (what it is declared as conceptually)
Runtime type (actual type): The real object stored in memory

```python
x = 10
"""Reference type: variable x
Runtime type: int object

In Python, variables do not fix types — objects do."""
```

# 4. Why constructor is split into new and init?

Because object creation has two steps:

__new__
Creates memory for object
Returns the object
__init__
Initializes object values

```python
obj = A()
# Internally
obj = A.__new__(A)
A.__init__(obj)
# So:

"""__new__ = creation
__init__ = setup"""
```

## 5. Can constructor be overridden or polymorphic?

Yes.

You can override __init__ in child class.
Python supports constructor polymorphism using default/variable arguments.

```python
class A:
    def __init__(self, x):
        self.x = x

class B(A):
    def __init__(self, x, y):
        super().__init__(x)
        self.y = y
# So constructor behaves differently in different classes.
```

## 6. Why constructor is not inherited?

Actually:
 __init__ is inherited BUT not automatically reused.


Because child class may need different initialization.

So Python allows override instead of forcing parent constructor behavior.

## 7. Why calling overridden method in constructor is risky?

```python
# Because object is not fully initialized yet.
class A:
    def __init__(self):
        self.show()

    def show(self):
        print("A")

class B(A):
    def show(self):
        print(self.x)  # may not exist yet
# If B() is created → error occurs.
"""
Reason:

constructor runs before full initialization of child class"""
```

8. Does Python support multiple inheritance? Why?

```python
"""Yes.

Because Python allows a class to inherit from more than one parent:"""
class A:
    pass

class B:
    pass

class C(A, B):
    pass
```

Why supported?

Python uses Method Resolution Order (MRO) to avoid conflicts
Flexible design choice (unlike Java)

## 9. What is MRO (Method Resolution Order)?

MRO defines the order in which Python searches methods in inheritance.

```python
class A: pass
class B(A): pass
class C(B): pass
# Order:

# C → B → A → object
"""Python uses C3 linearization algorithm to maintain consistency."""
```

## 10. What is super() actually doing?

super() refers to the parent class.

It:

Finds next class in MRO
Calls parent method without naming parent directly

```python
# super().__init__()
# Go to next class in MRO and call its method
```
