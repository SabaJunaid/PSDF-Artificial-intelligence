```python
departments = ("CS", "SE", "AI", "DS", "IT")
```

```python
passed_students = []
highest_marks = -1
top_student = ""

while True:
    name = input("Enter student name (or exit): ")

    if name.lower() == "exit":
        break

    dept = input("Enter department: ")
    marks = int(input("Enter marks: "))
    attendance = int(input("Enter attendance %: "))

    total_students += 1
```

```python
    if dept not in departments:
        print("Invalid Department")

    if marks >= passing_marks and attendance > 75:
        print("Eligible for Exam")
        passed_students.append(name)

    if marks > 85 and attendance > 80:
        print("Scholarship Candidate")

    if attendance < 50 or marks < 50:
        print("Student Not eligible")

    if marks > highest_marks:
        highest_marks = marks
        top_student = name
```
