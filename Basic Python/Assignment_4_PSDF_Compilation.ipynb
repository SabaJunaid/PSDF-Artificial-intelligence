```python
rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

## Question Shopping Cart and Login Management System


Q1. Login System (Dictionary + If-Else + Loop)

```python
users = {
    "admin": "1234",
    "ali": "1111"
}

attempts = 3

while attempts > 0:
    username = input("Enter username: ").strip()
    password = input("Enter password: ").strip()

    if username in users and users[username] == password:
        print("Login Successful")
        break
    else:
        attempts -= 1
        print("Invalid login. Attempts left:", attempts)

if attempts == 0:
    print("Account Locked")
    exit()
```

Q2. Shopping Menu (Products Dictionary)

```python
products = {
    "Apple": 100,
    "Banana": 40,
    "Mango": 120,
    "Orange": 80
}

print("Available Products:")
for i, j in products.items():
    print(i, ":", j)
```

Q3. Add Products to Cart (List + Dictionary)

```python
cart = {}

while True:
    item = input("Enter product (or done): ").strip()

    if item == "done":
        break

    if item in products:
        qty = int(input("Enter quantity: "))

        if item in cart:
            cart[item] += qty
        else:
            cart[item] = qty
    else:
        print("Product not found")
```

Q4. Update / Remove Products

```python
action = input("Update or Remove? ").strip()

if action == "update":
    item = input("Product name: ")
    if item in cart:
        qty = int(input("New quantity: "))
        cart[item] = qty

elif action == "remove":
    item = input("Product name: ")
    if item in cart:
        del cart[item]
```

Q5. Total Bill + Most Expensive Product

```python
total = 0
max_price = 0
max_item = ""

for item, qty in cart.items():
    price = products[item] * qty
    total += price

    if price > max_price:
        max_price = price
        max_item = item

print("Total Bill:", total)
print("Most Expensive Item:", max_item)
```
