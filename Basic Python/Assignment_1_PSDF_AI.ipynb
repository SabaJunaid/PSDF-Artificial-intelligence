## Q1. Store and print Name, Age, City

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
city = input("Enter your city: ")

print(name)
print(age)
print(city)
```

## Q2. Positive, Negative, or Zero

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

## Q3. Even or Odd

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

## Q4. Pass or Fail (Marks)

```python
marks = int(input("Enter marks: "))

if marks >= 50:
    print("Pass")
else:
    print("Fail")
```

## Q5. Voting Eligibility

```python
age = int(input("Enter age: "))

if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

## Q6. Temperature Checker

```python
temp = int(input("Enter temperature: "))

if temp > 35:
    print("Hot")
elif 20 <= temp <= 35:
    print("Normal")
else:
    print("Cold")
```

## Q7. Student Information System

```python
name = input("Enter name: ")
age = int(input("Enter age: "))
marks = float(input("Enter marks: "))
enrolled_input = input("Is student enrolled? (True/False): ")

# convert string to boolean
if enrolled_input == "True":
    enrolled = True
else:
    enrolled = False

# Condition checks
if marks >= 50 and age >= 18:
    result = "PASS and eligible for higher studies"

elif marks >= 80 or enrolled == True:
    result = "EXCELLENT / Active Participant"

else:
    result = "Student needs improvement"

# Final output
print("----- Student Record -----")
print("Name:", name)
print("Age:", age)
print("Marks:", marks)
print("Enrolled:", enrolled)
print("Result:", result)
```
