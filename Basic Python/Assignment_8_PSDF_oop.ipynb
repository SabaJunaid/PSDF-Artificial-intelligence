1. Library Book Management System
Create a class named LibraryBook.
Requirements:
• Attributes:
o title (public)
o author (public)
o _availableCopies (protected)
• Constructor should initialize all attributes.
• Methods:
o borrowBook() → decrease available copies by 1 if copies are available.
o returnBook() → increase available copies by 1.
o showBookDetails() → display all book information.
• Create at least 2 book objects and perform borrowing and returning operations.

```python
class LibraryBook:
    def __init__(self, title, author, availableCopies):
        self.title = title
        self.author = author
        self._availableCopies = availableCopies  # protected attribute

    def borrowBook(self):
        if self._availableCopies > 0:
            self._availableCopies -= 1
            print(f"You borrowed '{self.title}'.")
        else:
            print(f"Sorry, '{self.title}' is not available.")

    def returnBook(self):
        self._availableCopies += 1
        print(f"You returned '{self.title}'.")

    def showBookDetails(self):
        print("Book Details:")
        print("Title:", self.title)
        print("Author:", self.author)
        print("Available Copies:", self._availableCopies)
        print("-" * 30)


# Creating 2 book objects
book1 = LibraryBook("The Alchemist", "Paulo Coelho", 3)
book2 = LibraryBook("Atomic Habits", "James Clear", 2)

# Performing operations on book1
book1.showBookDetails()
book1.borrowBook()
book1.borrowBook()
book1.returnBook()
book1.showBookDetails()

# Performing operations on book2
book2.showBookDetails()
book2.borrowBook()
book2.borrowBook()
book2.borrowBook()  # testing limit case
book2.returnBook()
book2.showBookDetails()
```

2. Employee Management System
Create a class named Employee.
Requirements:
• Attributes:
o name (public)
o designation (public)
o _salary (protected)
• Constructor should initialize all attributes.
• Methods:
o showDetails() → display employee information.
o increaseSalary(amount) → increase salary by the given amount.
o showSalary() → display the current salary.
• Create 3 employee objects and give a salary in

```python
class Employee:
    def __init__(self, name, designation, salary):
        # Constructor initializes employee data
        self.name = name                  # public attribute
        self.designation = designation    # public attribute
        self._salary = salary             # protected attribute

    def showDetails(self):
        # Displays employee basic information
        print("Employee Details:")
        print("Name:", self.name)
        print("Designation:", self.designation)
        print("-" * 25)

    def increaseSalary(self, amount):
        # Increases salary by given amount
        self._salary += amount
        print(f"Salary increased by {amount} for {self.name}")

    def showSalary(self):
        # Displays current salary
        print(f"Current salary of {self.name}: {self._salary}")


# Creating 3 employee objects
emp1 = Employee("Ali", "Manager", 50000)
emp2 = Employee("Sara", "Developer", 40000)
emp3 = Employee("Ahmed", "Designer", 35000)

# Showing details
emp1.showDetails()
emp1.showSalary()

emp2.showDetails()
emp2.showSalary()

emp3.showDetails()
emp3.showSalary()

# Giving salary increment to one employee
emp2.increaseSalary(5000)

# Showing updated salary
emp2.showSalary()
```
