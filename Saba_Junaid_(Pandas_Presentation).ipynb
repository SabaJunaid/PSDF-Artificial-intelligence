## Why do we use dropna()?

##  **3. str.contains()**

## What is str.contains()?

str.contains() is a pandas string method used to check whether a particular text pattern exists inside strings of a Series (usually a DataFrame column).

It returns True if the pattern is found and False if it is not found.

## Why is str.contains() needed?

Suppose you have 100,000 records.

You want to find:

* All students whose names contain "Ali"
* All emails containing "@gmail.com"
* All products containing "Laptop"
* All cities containing "Lahore"

Checking manually is impossible.

str.contains() automatically searches every row and tells us where the pattern exists.

# What problem does it solve?

Without str.contains():

**for value in column:**
    
    # manually search text


This is slow and complicated.

str.contains() performs vectorized searching on the entire column at once.

```python
# Real-life Example 1: Student Names

students = pd.DataFrame({
    "Name": ["Ali Ahmed", "Sara Khan", "Ali Raza", "Ayesha"]
})

print(students)
```

```python
## Find names containing "Ali"
students["Name"].str.contains("Ali")

# We want to identify rows containing "Ali".

#  True means found.

# False means not found.
```

```python
# for filtering Rows Usually we use:

students[
    students["Name"].str.contains("Ali")
]
```

```python
# Email Analysis

emails = pd.DataFrame({
    "Email":[
        "ali@gmail.com",
        "sara@yahoo.com",
        "ahmed@gmail.com"
    ]
})
```

```python
# Find Gmail users:

emails[
emails["Email"].str.contains("@gmail")
]
```

Why is this useful?

Companies often separate customers based on:

* Email providers
* Domains
* Countries
* Product categories

```python
"""Series.str.contains(
pat,
case=True,
flags=0,
na=None,
regex=True
)"""
```

| Parameter | Explanation                                                                                                                                                         |
| --------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `pat`     | The text or pattern you want to search for inside strings. Without a search pattern, Pandas does not know what to find.                                             |
| `case`    | Controls case sensitivity. If `True`, the search is case-sensitive; if `False`, it ignores uppercase/lowercase differences.                                         |
| `flags`   | Used when working with advanced regular expression (regex) patterns. It allows the use of regular expression flags (e.g., `IGNORECASE` to ignore capitalization).   |
| `na`      | Defines what to return for missing values (`NaN`). Without it, missing values may create errors or unwanted `NaN` results.                                          |
| `regex`   | Determines whether the pattern is treated as a regular expression. If `True`, `pat` is interpreted as a regex pattern; if `False`, it searches for exact text only. |


## Regex
(Regular Expression) is a special pattern used to search, match, or filter text.

| Regex   | Meaning              | Matches    |
| ------- | -------------------- | ---------- |
| `"Ali"` | Contains Ali         | Ali Ahmed  |
| `"^A"`  | Starts with A        | Ali, Ahmed |
| `"a$"`  | Ends with a          | Sara       |
| `"\d"`  | Contains a number    | Ali123     |
| `"."`   | Any single character | A, B, 5, @ |

## **4. melt() in Pandas**

## What is melt()?

melt() is a pandas function used to convert wide-format data into long-format data.

It reshapes a DataFrame by turning columns into rows.

## Why is melt() needed?

Many datasets are stored in a wide format because humans find them easy to read.

Example:
| Student | Math | English | Science |
| ------- | ---- | ------- | ------- |
| Ali     | 80   | 75      | 90      |
| Sara    | 85   | 88      | 92      |

But statistical analysis and visualization often require long format:

| Student | Subject | Marks |
| ------- | ------- | ----- |
| Ali     | Math    | 80    |
| Ali     | English | 75    |
| Ali     | Science | 90    |




## Real-life Example: Student Marks

```python
df = pd.DataFrame({
"Student":["Ali","Sara"],
"Math":[80,85],
"English":[75,88],
"Science":[90,92]
})

print(df)
```

```python
# Applying melt()
long_df = pd.melt(
df,
id_vars=["Student"],
value_vars=["Math","English","Science"],
var_name="Subject",
value_name="Marks"
)

print(long_df)
```

# Why use melt() here?

Because subjects were stored as separate columns. We converted them into one Subject column, making analysis easier.

```python
"""pd.melt(
frame,
id_vars,
value_vars,
var_name,
value_name
)"""
```

| Parameter    | Explanation                                                                                   | Why Needed?                                                                                                                    | Example                                 |
| ------------ | --------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------ | --------------------------------------- |
| `frame`      | The DataFrame that you want to reshape from wide format to long format.                       | Pandas needs to know which DataFrame should be transformed.                                                                    | `python pd.melt(df) `                   |
| `id_vars`    | Columns that should remain unchanged during transformation. These columns act as identifiers. | We use it because some columns contain important information (like Student ID or Name) that should not be converted into rows. | `python id_vars=["Student"] `           |
| `value_vars` | Columns that will be converted from columns into rows.                                        | It tells Pandas which columns need to be reshaped. If not provided, all columns are melted except `id_vars`.                   | `python value_vars=["Math","English"] ` |
| `var_name`   | New name given to the column that contains the original column names after melting.           | It makes the reshaped data easier to understand by replacing the default column name `"variable"`.                             | `python var_name="Subject" `            |
| `value_name` | New name given to the column that contains the values from the original columns.              | It replaces the default name `"value"` and gives a meaningful name to the data values.                                         | `python value_name="Marks" `            |


## **Alternatives of melt()**

## **6. insert() in Pandas**

## What is insert()?

insert() is a pandas DataFrame method used to add a new column at a specific position in a DataFrame.

Unlike normal column assignment, which always adds a column at the end, insert() allows us to decide exactly where the new column should appear.

## Why is insert() needed?

Suppose you have an employee dataset:
| Name | Salary |
| ---- | ------ |
| Ali  | 50000  |
| Sara | 60000  |

Now you want to add Department between Name and Salary:

Desired:

| Name | Department | Salary |
| ---- | ---------- | ------ |
| Ali  | IT         | 50000  |

## If we use:

df["Department"] = ["IT","HR"]
The new column will automatically appear at the end:

Name	Salary	Department
Ali	50000	IT

This may not be suitable for reports.

insert() solves this problem by allowing controlled column placement.

## Real-life Example

A company has employee data.

Initially:

**Employee_ID | Name | Salary**

After collecting new information, we want:

**Employee_ID | Name | Department | Salary**

The Department column should be placed before Salary.


## Purpose of insert()

The main purposes are:

* Adding a new column.
* Controlling column order.
* Preparing datasets for reports and analysis.
* Organizing data in a meaningful structure.

```python
"""Syntax
DataFrame.insert(
    loc,
    column,
    value,
    allow_duplicates=False
)"""
```

| Parameter          | Explanation                                                                                                 | Why Needed?                                                                                                                                 |
| ------------------ | ----------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------- |
| `loc`              | Defines the position (index number) where the new column will be inserted. Column positions start from `0`. | Pandas needs to know where the new column should appear in the DataFrame.                                                                   |
| `column`           | Specifies the name of the new column that will be created.                                                  | Every column needs a label/name so it can be accessed and identified later.                                                                 |
| `value`            | Contains the data that will be stored in the new column.                                                    | Without values, Pandas cannot create and fill the new column.                                                                               |
| `allow_duplicates` | Controls whether duplicate column names are allowed. The default value is `False`.                          | It prevents accidentally creating columns with the same name. Set it to `True` only when duplicate column names are intentionally required. |

**7. at[] in Pandas**

## What is at[]?

at[] is a pandas accessor used to access or modify a single value in a DataFrame using row and column labels.

It is designed specifically for fast single-cell operations.

## Why is at[] needed?

Sometimes we do not want to select an entire row or column. We only need one specific value.

Example:

A student's marks were entered incorrectly.

Original:

| Name | Marks |
| ---- | ----- |
| Ali  | 80    |


Need to change only Ali's marks to 90.

Using normal selection methods is unnecessary.

at[] directly targets that cell.

## Purpose of at[]
* Reading one value.
* Updating one value.
* Fast single-cell access.

## Real-life Example

Updating one customer's phone number in a database.


```python
# Syntax
# DataFrame.at[row_label, column_label]
```

```python
import pandas as pd

students=pd.DataFrame({
"Name":["Ali","Sara"],
"Marks":[80,90]
})

print(students)
```

```python
# Accessing a value
students.at[0,"Marks"]
```

```python
# Modifying a value
students.at[0,"Marks"]=95

print(students)
```

## Why use at() instead of loc()?

Both can access values.

Example:

**df.loc[0,"Marks"]**

also works.

But:

at[] → optimized for one value.

loc[] → designed for selecting multiple rows and columns.

Examples:

* Stock prices
* Weather measurements
* Sales records
* Sensor readings

Analysts frequently need recent data.

Instead of manually filtering dates, last() quickly extracts the required recent period.

To retrieve data from the end of a time-based dataset.

A company records daily sales
| Date  | Sales |
| ----- | ----- |

Manager wants the last 2 days' sales.

last() retrieves them quickly.

## **9. pct_change() in Pandas**

## What is pct_change()?

pct_change() is a pandas method used to calculate the percentage change between the current value and the previous value in a Series or DataFrame.

It tells us how much a value has increased or decreased compared with its previous value.

The result is returned as a decimal value.

Example:

If sales increase from 100 to 120:


120−100/100
	​

=0.20

Output:

0.20

which means 20% increase.

## Why is pct_change() needed?

Looking only at actual values does not always show the growth pattern.

Example:

Monthly sales:

| Month    | Sales |
| -------- | ----- |
| January  | 1000  |
| February | 1500  |
| March    | 1800  |


## A manager wants to know:

* How much did sales grow?
* Which month had the highest increase?
* Is growth improving or decreasing?

## The difference alone:

January → February:

1500 - 1000 = 500

does not tell the growth rate.

## Percentage change gives:

50% increase

which is easier to compare.

## **Purpose of pct_change()**

The main purposes are:

* Measuring growth rate.
* Comparing changes over time.
* Financial analysis.
* Sales analysis.
* Economic data analysis.
8 Time-series analysis.

```python
## A company records monthly sales.

sales = pd.DataFrame({
    "Month":["January","February","March","April"],
    "Sales":[1000,1500,1200,1800]
})

print(sales)
```

```python
# Applying pct_change()
sales["Growth"] = sales["Sales"].pct_change()

print(sales)
```

## Explanation of Output

**First row:**

NaN

because there is no previous value to compare January with.

**February:**

Previous sales = 1000

Current sales = 1500

Calculation:

(1500-1000)/1000

=0.5

Meaning:

50% increase

**March:**

(1200-1500)/1500

=-0.20

Meaning:

20% decrease

**April:**

(1800-1200)/1200

=0.50

Meaning:

50% increase

## **Converting decimal into percentage**

Normally we multiply by 100:

sales["Growth_percentage"] = (

)

## **10. to_stata() in Pandas**

**What is to_stata()?**

to_stata() is a pandas method used to export a DataFrame into a Stata data file format (.dta).

It allows data created or cleaned in pandas to be opened and analyzed in Stata software.

**Why is to_stata() needed?**

Different researchers and organizations use different statistical software.

For example:

A data scientist may clean data using Python/pandas.
A statistician may perform statistical analysis using Stata.

They need a way to transfer data between programs.

to_stata() provides this connection.

**What is Stata?**

Stata is a statistical software widely used in:

* Economics
* Public health
* Social sciences
* Research studies

It uses its own file format:

.dta

**Purpose of to_stata()**

The purpose is:

* Export pandas DataFrames.
* Share data with Stata users.
* Maintain data structure during transfer.

```python
# Syntax
"""DataFrame.to_stata(
path,
write_index=True,
version=None
)"""
```

| Parameter     | Explanation                                                                                                                             | Why Needed?                                                                                                                                           |
| ------------- | --------------------------------------------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------- |
| `path`        | Specifies the location and name of the Stata file where the DataFrame will be saved. The file is usually saved with a `.dta` extension. | Pandas needs to know where to store the exported data. Without a path, it cannot create the Stata file.                                               |
| `write_index` | Determines whether the DataFrame index should be written into the Stata file. Default value is `True`.                                  | Controls whether row labels (index values) are included in the exported file. If the index is not useful, it can be removed by setting it to `False`. |
| `version`     | Specifies the version of the Stata file format to create.                                                                               | Different versions of Stata support different features and data sizes. This ensures compatibility with the required Stata software version.           |
